## 3.1. Correlation Matrix

```python
# Drop categorical columns
df_matrix = df_quality.drop(columns = ['Species', 'Continent_of_Origin', 'Country_of_Origin', 'Harvest_Year',
       'Expiration', 'Variety', 'Color', 'Processing_Method','Quakers', 'Category_One_Defects',
       'Category_Two_Defects'])

# Correlation matrix
df_matrix.corr()
```

## 3.2. HeatMap

```python
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(10, 6))

# Generate the correlation matrix
corr_matrix = df_matrix.corr()

# Create the heatmap with annotations and a labelled colour bar
sns.heatmap(corr_matrix,
            annot=True,
            cbar_kws={'label': 'Correlation Coefficient'},
            ax=ax)

# Add labels and title
ax.set_title("Correlation Matrix - Coffee Quality Measures")
ax.set_xlabel("Measures")
ax.set_ylabel("Measures")

# Adjust layout
plt.tight_layout()

# Save the figure before displaying
save_path = os.path.join(path, 'visualisations', 'eda_02_correlation_matrix.png')
plt.savefig(save_path, dpi=300, bbox_inches='tight')

# Show the plot
plt.show()

# Print confirmation message
print(f"Chart successfully saved at: {save_path}")
```

##### Measure Group 1: 'Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance'
- The correlation coefficients among measures in this group are all above +0.5, indicating **strong positive relationships**. 
- Notably, 'Aftertaste' show a very strong correlation with 'Flavor' and 'Balance' of +0.83 and +0.75 respectively.

##### Measure Group 2: 'Uniformity', 'Clean_Cup', 'Sweetness'
- Correlation coefficients within this group range between +0.43 and +0.49, reflecting **moderate positive relationships**.

##### Group 1 vs Group 2: 
- Correlation coefficients between measures in Group 1 and Group 2 generally fall between 0 and +0.3, suggesting **weak positive relationships** across groups.

##### Measure: 'Moisture'
- Correlation with Group 1 measures is weakly **negative** (ranging from -0.15 to -0.05).
- Correlation with Group 2 measures is **very weakly positive**, all between 0 and +0.05.

## 3.3. Pair plot

```python
# Create subset with sensory evaluations
df_pair_plot = df_quality[['Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance','Uniformity', 'Clean_Cup', 'Sweetness', 'Moisture']]

# Create the pair plot
pair_plot = sns.pairplot(df_pair_plot)

# Add a title manually (not supported directly via seaborn API)
pair_plot.fig.suptitle("Relationships Between Coffee Sensory Attributes")
pair_plot.fig.tight_layout()
pair_plot.fig.subplots_adjust(top=0.96)  # Adjust to make room for the title

# Define export path
save_path = os.path.join(path, 'visualisations', 'eda_03_pairplot.png')

# Ensure directory exists
os.makedirs(os.path.dirname(save_path), exist_ok=True)

# Save the figure
pair_plot.savefig(save_path, dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

##### Measure Group 1: 'Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance'
- Linear relationships with a strong correlation.
- Normal distribution.
- Suitable for further exploration through regression analysis, for example.


Group 1: Aroma, Flavor, Aftertaste, Acidity, Body, Balance
- The scatterplots among these variables form tight, upward-trending clusters, visually supporting the strong positive correlations (r > 0.5) reported in the heatmap.
- The relationships involving Aftertaste â€” especially with Flavor and Balance â€” are particularly linear and dense, aligning with the highest coefficients of +0.83 and +0.75.
- Other variable pairs in this group (e.g., Aroma vs Flavor, Acidity vs Balance) show consistent upward trends, albeit with slightly more spread, confirming a strong but not perfect association.

##### Measure Group 2: 'Uniformity', 'Clean_Cup', 'Sweetness'
- Non-linear relationships which form clusters along specific lines.
- Heavily left-skewed, with most scores concentrated at the top of the scale.
- Possible scoring bias.
- Explore alternative ways to investigate these relationships.

##### Variables: 'Moisture'

```python
# Copy dataframe
df_categorical = df_quality.copy()

# Create loop
result = []
for value in df_categorical['Flavor']:
    if value < 7.25:
        result.append('below average')
    elif 7.25 <= value <= 7.75:
        result.append('average')
    else:
        result.append('high')

# Create new variable
df_categorical['Flavor_category'] = result
```

```python
# Frequency - new variable
df_categorical['Flavor_category'].value_counts(dropna = False)
```

```python
# Set style
sns.set(style="ticks")

# Create figure
plt.figure(figsize=(10, 6))

# Scatterplot
cat_scat_plot = sns.scatterplot(
    x="Acidity",
    y="Flavor",
    hue="Flavor_category",
    data=df_categorical
)

# Add labels and title
plt.title("Acidity vs. Flavor by Flavour Category")
plt.xlabel("Acidity Score")
plt.ylabel("Flavour Score")

# Improve legend title
plt.legend(title="Flavor Category", title_fontsize='10', fontsize='9')

# Adjust layout
plt.tight_layout()

# Define export path
save_path = os.path.join(path, 'visualisations', 'eda_06_category_acidity_flavour.png')

# Ensure directory exists
os.makedirs(os.path.dirname(save_path), exist_ok=True)

# Save the figure
plt.savefig(save_path, dpi=300, bbox_inches='tight')

# Show the plot
plt.show()

# Print confirmation
print(f"Chart successfully saved at: {save_path}")
```

##### Previous insights
- Strong positive correlation.
- Many points clustered near the line indicating low degree of variance.
- Similar outlier pattern with more below the line, but one lies above the line at a low Acidity score, while Flavor scores high.

##### Further comments
- Flavor category supports the correlation, with higher flavor scores at the higher range of acidity scores.
- Consistency suggests predictability.
- One outlier could be due to unique characteristics, or an error.<br><br>
